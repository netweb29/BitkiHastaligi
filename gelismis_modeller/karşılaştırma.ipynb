```python
import os
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from PIL import Image
import random
import matplotlib.pyplot as plt
import numpy as np
import urllib.parse

# ==============================================================================
# AYARLAR
# ==============================================================================
CSV_DOSYASI = r"C:\Users\FERHAT\Downloads\egitim_icin_hazir.csv"
ANA_KLASOR = r"C:\Users\FERHAT\Plant\New Plant Diseases Dataset(Augmented)\New Plant Diseases Dataset(Augmented)\train"
ALEXNET_PATH = r"C:\Users\FERHAT\alexnet_model_final.pth"
SQUEEZENET_PATH = r"C:\Users\FERHAT\squeezenet_model_final.pth"
RESIM_BOYUTU = 224
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ==============================================================================
# VERI SETI SINIFI
# ==============================================================================
class TankVeriSeti(Dataset):
    def __init__(self, csv_file, root_dir, transform=None):
        self.veriler = pd.read_csv(csv_file)
        self.root_dir = root_dir
        self.transform = transform
        self.sinif_isimleri = sorted(self.veriler['class'].unique())
        self.sinif_sozlugu = {isim: i for i, isim in enumerate(self.sinif_isimleri)}
        self.idx_to_class = {i: isim for isim, i in self.sinif_sozlugu.items()}
        
        self.dosya_haritasi = {}
        for root, dirs, files in os.walk(root_dir):
            for file in files:
                tam_yol = os.path.join(root, file)
                self.dosya_haritasi[file] = tam_yol
                duzgun_isim = urllib.parse.unquote(file)
                if duzgun_isim != file:
                    self.dosya_haritasi[duzgun_isim] = tam_yol

    def __len__(self): return len(self.veriler)

    def __getitem__(self, idx):
        try:
            dosya_adi = str(self.veriler.iloc[idx, 0]).split('?')[0]
            hedef_yol = self.dosya_haritasi.get(dosya_adi)
            if not hedef_yol:
                hedef_yol = self.dosya_haritasi.get(urllib.parse.unquote(dosya_adi))
            if not hedef_yol:
                return self.__getitem__(random.randint(0, len(self.veriler)-1))

            image = Image.open(hedef_yol).convert("RGB")
            label = self.sinif_sozlugu[str(self.veriler.iloc[idx, 1])]
            if self.transform: image = self.transform(image)
            return image, label
        except:
            return self.__getitem__(random.randint(0, len(self.veriler)-1))

# ==============================================================================
# MODEL YUKLEME
# ==============================================================================
def model_yukle(isim, yol, device):
    if isim == "AlexNet":
        model = models.alexnet(weights=None)
        model.classifier[6] = nn.Linear(4096, 43)
    elif isim == "SqueezeNet":
        model = models.squeezenet1_0(weights=None)
        model.classifier[1] = nn.Conv2d(512, 43, kernel_size=(1,1))
        model.num_classes = 43
    
    model = model.to(device)
    
    if os.path.exists(yol):
        state_dict = torch.load(yol, map_location=device)
        new_state_dict = {k.replace("model.", "") if k.startswith("model.") else k: v for k, v in state_dict.items()}
        try:
            model.load_state_dict(new_state_dict)
        except:
            pass
    model.eval()
    return model

# ==============================================================================
# METIN TEMIZLEME FONKSIYONU
# ==============================================================================
def temizle_ve_kisalt(metin, limit=20):
    # Alt cizgileri bosluk yap
    temiz = metin.replace("_", " ").replace("Tomato", "").replace("Potato", "")
    temiz = temiz.strip()
    # Cok uzunsa kisalt
    if len(temiz) > limit:
        return temiz[:limit] + "..."
    return temiz

# ==============================================================================
# KARSILASTIRMA VE CIZIM
# ==============================================================================
def karsilastir_ve_ciz(alex_model, sqz_model, loader, class_map, device):
    images, labels = next(iter(loader))
    images, labels = images.to(device), labels.to(device)
    
    with torch.no_grad():
        out_alex = alex_model(images)
        _, pred_alex = torch.max(out_alex, 1)
        
        out_sqz = sqz_model(images)
        _, pred_sqz = torch.max(out_sqz, 1)
    
    total = labels.size(0)
    acc_alex = (pred_alex == labels).sum().item() / total * 100
    acc_sqz = (pred_sqz == labels).sum().item() / total * 100

    # Grafik Boyutu
    fig = plt.figure(figsize=(22, 24)) 
    
    plt.suptitle(f"MODEL KARILATIRMASI\nAlexNet: %{acc_alex:.1f}  |  SqueezeNet: %{acc_sqz:.1f}", 
                 fontsize=24, fontweight='bold', y=0.96)
    
    num_img = min(len(images), 12)
    
    # Izgara yerlesimi 
    plt.subplots_adjust(hspace=0.6, wspace=0.3)

    for i in range(num_img):
        ax = plt.subplot(4, 3, i+1) # 4 satir, 3 sutun yaptik (Daha buyuk gorunur)
        
        img = images[i].cpu().permute(1, 2, 0).numpy()
        img = img * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
        img = np.clip(img, 0, 1)
        
        ax.imshow(img)
        ax.axis('off')
        
        # Metinler
        gercek_ham = class_map.get(labels[i].item(), "Bilinmiyor")
        alex_ham = class_map.get(pred_alex[i].item(), "Bilinmiyor")
        sqz_ham = class_map.get(pred_sqz[i].item(), "Bilinmiyor")
        
        gercek_yazi = temizle_ve_kisalt(gercek_ham)
        alex_yazi = temizle_ve_kisalt(alex_ham)
        sqz_yazi = temizle_ve_kisalt(sqz_ham)
        
        # Renk Belirleme
        c_alex = 'green' if pred_alex[i] == labels[i] else 'red'
        c_sqz = 'green' if pred_sqz[i] == labels[i] else 'red'
        
        # 1. UST KISIM: Gercek Sinif
        ax.set_title(f"GEREK: {gercek_yazi}", fontsize=14, fontweight='bold', pad=10, color='black')
        
        # 2. ALT KISIM: Modellerin Tahminleri (Ayri satirlar)
        # AlexNet Yazisi
        ax.text(0.5, -0.12, f"AlexNet: {alex_yazi}", 
                transform=ax.transAxes, ha='center', fontsize=12, color=c_alex, fontweight='bold')
        
        # SqueezeNet Yazisi
        ax.text(0.5, -0.22, f"SqzNet: {sqz_yazi}", 
                transform=ax.transAxes, ha='center', fontsize=12, color=c_sqz, fontweight='bold')

    plt.savefig("karsilastirma_duzgun.png", bbox_inches='tight')
    plt.show()
    print("Gorsel kaydedildi: karsilastirma_duzgun.png")

# ==============================================================================
# CALISTIR
# ==============================================================================
if __name__ == "__main__":
    transform = transforms.Compose([
        transforms.Resize((RESIM_BOYUTU, RESIM_BOYUTU)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    try:
        ds = TankVeriSeti(CSV_DOSYASI, ANA_KLASOR, transform)
        loader = DataLoader(ds, batch_size=12, shuffle=True)
        
        alex = model_yukle("AlexNet", ALEXNET_PATH, DEVICE)
        sqz = model_yukle("SqueezeNet", SQUEEZENET_PATH, DEVICE)
        
        karsilastir_ve_ciz(alex, sqz, loader, ds.idx_to_class, DEVICE)
        
    except Exception as e:
        print(f"Hata: {e}")
```
